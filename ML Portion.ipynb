```python
# Read the file into a pandas DataFrame

hurr = pd.read_csv('Historical_Tropical_Storm_Tracks.csv')
# hurr.head()
# hurr.dtypes
```

```python
# drop unwanted columns
# Drop all years >1950, drop all Pacific occurences, drop name, BTID
# create new dataframe

# hurr.dtypes

hurr.sort_values(by=['YEAR'], inplace=True, ascending=True)
# hurr.tail
hurr.drop(hurr[hurr['YEAR'] < 1950].index, inplace = True)
hurr.drop(hurr[hurr['CAT'] < "H3"].index, inplace = True)
hurr.drop(hurr[hurr['BASIN'] == 'Eastern Pacific'].index, inplace=True)
hurr.drop(['FID', 'BTID','Shape_Leng','AD_TIME'], axis=1)
```
