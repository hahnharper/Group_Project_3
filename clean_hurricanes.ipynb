```python
from datetime import datetime, timedelta
import time
from collections import namedtuple
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The path to our CSV file
hurricanes = "Historical_Tropical_Storm_Tracks.csv"


# Read our hurricane data into pandas
hurricanes_df = pd.read_csv(hurricanes)
hurricanes_df.head()
```

```python

hurricanes_df = hurricanes_df.dropna(how='any')
hurricanes_df.head()
```

```python
hurricanes_df = hurricanes_df.dropna(how='any')
hurricanes_df.head()
```

```python
hurricanes_df.sort_values(by=['YEAR'], inplace=True, ascending=True)
# hurr.tail
hurricanes_df.drop(hurricanes_df[hurricanes_df['YEAR'] < 1950].index, inplace = True)
#hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] < "H3"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "TS"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "TD"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "SD"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "SS"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "W"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "L"].index, inplace = True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['BASIN'] == 'Eastern Pacific'].index, inplace=True)
hurricanes_df.drop(hurricanes_df[hurricanes_df['CAT'] == "E"].index, inplace = True)
hurricanes_df.drop(['FID', 'BTID','Shape_Leng','AD_TIME'], axis=1)
```
