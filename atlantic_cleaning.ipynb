```python
from datetime import datetime, timedelta
import time
from collections import namedtuple
import pandas as pd
import requests
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The path to our CSV file
atl_hurricanes = "atlantic.csv"


# Read our hurricane data into pandas
atl_hurricanes_df = pd.read_csv(atl_hurricanes)
atl_hurricanes_df.head()
```

```python
atl_hurricanes_df.sort_values(by=['Date'], inplace=True, ascending=True)
atl_hurricanes_df.drop(atl_hurricanes_df[atl_hurricanes_df['Date'] < 19500101].index, inplace = True)
atl_hurricanes_df.drop(atl_hurricanes_df[atl_hurricanes_df['Name'] == "UNNAMED"].index, inplace = True)
atl_hurricanes_df.drop(atl_hurricanes_df[atl_hurricanes_df['Status'] == "TS"].index, inplace = True)
atl_hurricanes_df.drop(atl_hurricanes_df[atl_hurricanes_df['Status'] == "TD"].index, inplace = True)
atl_hurricanes_df.drop(atl_hurricanes_df[atl_hurricanes_df['Maximum Wind'] < 75].index, inplace = True)
del atl_hurricanes_df['ID']
del atl_hurricanes_df['Event']
del atl_hurricanes_df['Minimum Pressure']
del atl_hurricanes_df['Low Wind NE']
del atl_hurricanes_df['Low Wind SE']
del atl_hurricanes_df['Low Wind SW']
del atl_hurricanes_df['Low Wind NW']
del atl_hurricanes_df['Moderate Wind NE']
del atl_hurricanes_df['Moderate Wind SE']
del atl_hurricanes_df['Moderate Wind NW']
del atl_hurricanes_df['High Wind NE']
#del atl_hurricanes_df['Moderate Wind SW ']
del atl_hurricanes_df['High Wind SW']
del atl_hurricanes_df['High Wind NW']
del atl_hurricanes_df['High Wind SE']

atl_hurricanes_df.head(100)
```
